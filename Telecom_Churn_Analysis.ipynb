### Donut Chart

```python
#data is taken from variablr A
#plt.pie(A)
plt.pie(A, labels=['Not churned', 'Churned'], colors=['pink','teal'], startangle = 50,
shadow = True, radius = 0.75, explode = (0,0.1), autopct = '%1.2f%%', pctdistance= 0.70)
#(data, label, colors, the start angle of the part, shadow, radius of the circle, explode separates the parts,
#auto percentage, distance of percentage from centre)

circle = plt.Circle((0,0), 0.35, color = 'white')
#creates circle of radius 0.35 in white colour

c = plt.gcf()
#retrieves the current figure (get current figure)

c.gca().add_artist(circle)
#adds the created white circle and places it in axes

plt.title('Churned vs Not Churned Customer')
plt.show()
```

### CountPlot

```python
plt.figure(figsize=(4,4.5))

chart = sns.countplot(x = 'churn', data = data, palette = ['pink','teal'])

total = len(data)

for p in chart.patches:
    count = p.get_height()
    percentage = f'{round((count/total)*100)}%'
    chart.annotate(percentage,(p.get_x() + p.get_width()/2, count),
                               ha ='center', va= 'bottom', fontsize=12, color = 'black')
plt.title('Churn Count')
# plt.xlabel('hi', color = 'purple')
plt.show()
```

### Count Plot

```python
sns.set (style = 'white')
plt.figure(figsize =(20,7))

cmap = LinearSegmentedColormap.from_list("pink_to_teal", ["pink", "teal"])
colors = [cmap(i/15) for i in range(15)]

chart = sns.countplot(x = 'state' , data =data,
                      order = data['state'].value_counts().head(15).index,
                      palette = colors)

total = len(data)

for p in chart.patches:
    percent = f'{round((p.get_height()/total)*100,1)}%'
    chart.annotate(percent, ((p.get_x() + p.get_width()/2), p.get_height()),
                  ha = 'center', va ='bottom', fontsize = 10)

plt.xticks(rotation = 45, ha = 'right')
#plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.title('Count of States with percentage (Sorted)', fontsize = 20)
plt.xlabel('States', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

### Count of churned and Unchurned statewise

```python
sns.set (style = 'white')
plt.figure(figsize = (20,7))

data['churn'] = data['churn'].astype(str)

color = {'False': 'pink', 'True' : 'teal'}
CS = sns.countplot(data = data, x = 'state', hue = 'churn',
              order = data['state'].value_counts().head(15).index, palette = color)

for p in CS.patches:
    CS.annotate(format(p.get_height(), '.0f'),
    (p.get_x()+p.get_width()/2., p.get_height()),
    ha = 'center', va = 'bottom',
    fontsize = 12, fontweight = 'bold')

CS.legend().set_title('Churn Status')

plt.title('Count of churn and unchurn statewise', fontsize = 20)
plt.xticks(rotation = 45, ha = 'right')
plt.xlabel('States', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```
